```python
import json
import os
```

```python
datasets = {}
data_dir = "../t_index_reproduce/data/synthetic/"
for lang in os.listdir(data_dir):
    lang_path = data_dir + lang + "/"
    datasets[lang] = {}
    for dataset in os.listdir(lang_path):
        ds_path = lang_path + dataset + "/"
        datasets[lang][dataset] = {}
        for split in os.listdir(ds_path):
            file_path = ds_path + split
            
            with open(file_path, "r") as file:
                raw_data = [json.loads(line) for line in file.readlines()]

            ds = []
            for i, item in enumerate(raw_data):
                ds.append({
                    "chosen": item['messages_foreignization']
                })
```

```python
datasets[lang][dataset][split.split('.')[0]] = []
```

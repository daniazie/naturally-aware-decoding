```python
import os
import json
```

```python
results_dir = "qa_decode"
results_files = os.listdir(results_dir)
```

```python
ntrex_dir = "../NTREX/NTREX-128"
```

```python
lang_map = {
    "ende": "deu",
    "enfr": "fra",
    "enzh": "zho-CN"
}

ntrex_prefix = "newstest2019-ref.{lang}.txt"
```

```python
results = {}
for result_file in results_files:
    with open(f"{results_dir}/{result_file}", "r") as file:
        results[result_file] = json.load(file)

    lang = result_file.split('_')[2]
    with open(f"{ntrex_dir}/{ntrex_prefix.format(lang=lang_map[lang])}", "r") as file:
        refs = file.readlines()

    for item, ref in zip(results[result_file], refs):
        item['ref'] = ref
```

```python
for result_file, result in results.items():
    with open(f"{results_dir}/{result_file}", "w") as file:
        json.dump(result, file, indent=2)
```

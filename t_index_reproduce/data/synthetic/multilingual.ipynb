```python
import json
import os
```

```python
lang_dirs = os.listdir()
```

```python
with open("enzh/mixture/train.jsonl", "r", encoding="utf-8") as file:
    mixture = []
    for line in file.readlines():
        mixture.append(json.loads(line))
```

```python
data = []
for lang_dir in lang_dirs:
    if not os.path.isdir(lang_dir):
        continue
    for dataset in os.listdir(lang_dir):
        if dataset == "mixture":
            continue
        data_path = lang_dir + "/" + dataset
        if "train.jsonl" in os.listdir(data_path):
            file_name = "train.jsonl"
        else:
            file_name = "test.jsonl"
        file_path = data_path + "/" + file_name
        with open(file_path, "r", encoding='utf-8') as file:
            for line in file.readlines()[:250]:
                data.append(json.loads(line))
```

```python
len(data)
```

```python
with open("train_multilingual.jsonl", "w", encoding="utf-8") as file:
    for item in data:
        json.dump(item, file)
```
